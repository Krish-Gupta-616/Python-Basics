6.  Write a function to demonstrate the behavior of mutable and immutable arguments

```python
#mutable
def modify_list(lst):
    lst.append(100)
    print("Inside function:", lst)
my_list = [1, 2, 3]
modify_list(my_list)
print("Outside function:", my_list)
```

```python
def modify_number(num):
    num = num + 10
    print("Inside function (immutable):", num)
x = 5
modify_number(x)
print("Outside function (immutable):", x)
```

```python
l = ['1',2,3,4,'5.7']
l1 = l.copy()
print(l)
k = 0
for i in l:
    i = float(i)
    l[k] = i 
    i = int(i)
    l1[k] = i
    k+= 1
print(l)
print(l1)
```

14.  Write a program to reverse a string using a while loop.

```python
a = input('Enter any string: ')
b = ''
k = 0
while k < len(a):
    b+= a[-1-k]
    k+= 1
print(b)
```
